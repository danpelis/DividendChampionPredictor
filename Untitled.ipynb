```python
import pandas as pd
import yfinance as yf
import datetime
import os
import numpy as np
```

```python
Sheets = dict()
for path,dirs,files in os.walk("./data/raw"):
    sort = np.argsort(files)
    year = path.split("/")[-1]
    if year not in ['raw','2021']:
        Sheets[year] = []
    for file in np.array(files)[sort]:
        Sheets[year].append(pd.read_excel(path+"/"+file, sheet_name=["Champions"], header=5))
```

```python
years = list(Sheets.keys())[1:]
```

```python
tickers = []

for year in years:
    for i in range(len(Sheets[year])):
        tickers.append(Sheets[year][i]['Champions']['Symbol'].values)
```
