```python
import pandas as pd
import yfinance as yf
import datetime
import os
import numpy as np
```

```python
Sheets = dict()
for path, dirs, files in os.walk("./data/raw"):
    sort = np.argsort(files)
    year = path.split("/")[-1]
    if year not in 'raw':
        Sheets[year] = []
    for file in np.array(files)[sort]:
        if file not in ".DS_Store":
            Sheets[year].append(pd.read_excel(path+"/"+file, sheet_name=["Champions"], header=5))
print(path)
```

```python
years = sorted(list(Sheets.keys())[1:])
years = years[1:]
```

```python
tickers = []

for year in years:
    for i in range(len(Sheets[year])):
        symbols = Sheets[year][i]['Champions']['Symbol']
        
        notna = pd.notna(symbols)
        symbols = symbols[notna].values.astype('<U5')
        
        notnum = np.char.isalpha(symbols)
        symbols = symbols[notnum]
        
        tickers.append(symbols)
```

```python
num_months = [len(Sheets[year]) for year in years]
num_months
```

```python
year_indx = [[years[i]]*num_months[i] for i in range(len(years))]
month_indx = [list(range(len(Sheets[year])+1))[1:] for year in years]
```

```python
year_indx = [item for sublist in year_indx for item in sublist]
month_indx = [item for sublist in month_indx for item in sublist]
```

```python
lost_status_date = []
lost_status = []

kept_status_date = []
kept_status = []

k=0
for i, ticker in enumerate(tickers):
    if k!= 0:
        prev = tickers[k-1]
        curr = tickers[k]
        lost = [sym not in curr for sym in prev]
        kept = [sym in curr for sym in prev]
        if list(prev[lost]):
            lost_status_date.append((year_indx[i-1], month_indx[i-1], prev[lost]))
        
        
    
```
